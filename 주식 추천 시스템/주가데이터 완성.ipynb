```python
import pandas as pd
import FinanceDataReader as fdr
```

```python
# 종목코드
code = '034730'

# 날짜
start_date = '2020-06-01'
end_date = '2024-05-31'

# 주가 데이터 수집
stock = fdr.DataReader(code, start=start_date, end=end_date).reset_index()

# 이동평균과 이격도 변수 추가
period = [5, 20, 60, 120]

for p in period :
    stock[f"{p}MA"] = stock["Close"].rolling(window = p).mean()
    stock[f"{p}Disparity"] = stock["Close"] / stock[f"{p}MA"]

print(stock)

stock.to_csv('./dataset/stock_sk.csv', index=False)
```

```python
# 감성분석 label 데이터 가져오기
final = pd.read_excel('./dataset/label_sk.xlsx')
final.rename(columns={'날짜':'Date'}, inplace=True)
final.drop('제목', axis=1, inplace=True)
final['Date'] = pd.to_datetime(final['Date'], format='%Y.%m.%d.')
final
```

```python
# 주가데이터
stock = pd.read_csv('./dataset/stock_sk.csv')
stock['Date'] = pd.to_datetime(stock['Date'])
stock
```

```python
# 뉴스기사 데이터는 평균내서 날짜를 기준으로 stock 데이터프레임과 합치기
grouped = final.groupby('Date')['label'].mean().reset_index()
merged_df = pd.merge(stock, grouped, on='Date', how='left')
merged_df
```

```python
# 시작 날짜와 종료 날짜를 설정
start = '2021-06-01'
end = '2024-05-31'

# 이 기간 동안의 데이터만 필터링
stock_df = merged_df[(merged_df['Date'] >= start) & (merged_df['Date'] <= end)]
stock_df
```

```python
stock_df.info()
```

```python
# 결측값 처리
# bfill로 결측값 채우기
stock_df.bfill(axis=0, inplace=True)  # 행 방향으로 결측값 채우기
# ffill로 결측값 채우기
stock_df.ffill(axis=0, inplace=True)  # 행 방향으로 결측값 채우기
stock_df.info()
```

```python
stock_df.to_csv(f'./dataset/final_sk.csv', index=False)
```
