```python
from transformers import pipeline
import pandas as pd
import tensorflow
```

```python
kor_clf_sentiment = pipeline("sentiment-analysis", "snunlp/KR-FinBert-SC")
```

```python
df = pd.read_excel('./dataset/naver.xlsx')
#df = pd.read_csv('./dataset/naver.csv')
df
```

```python
df_copy=df.copy()
df_copy
```

```python
# 제목 열에서 중복 확인
df_copy['제목'].duplicated().value_counts()
```

### 라벨링 진행

```python
df_copy['label'] = df_copy['제목'].apply(lambda x: kor_clf_sentiment(x)[0]['label'])
df_copy
```

```python
# 라벨을 숫자로 바꾸기
df_copy['label'].replace({'positive':1, 'neutral':0,'negative':-1 }, inplace=True)
df_copy
```

```python
df_copy['label'].value_counts()
```

```python
df_copy.to_excel('./dataset/label_naver.xlsx', index=False)
```
